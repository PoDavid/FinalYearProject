```python
#Only used to visiualize the join operation of the 4 files about disconnection
#Only one log is used to generate the file, thus only one log is handled
import pandas as pd
import os
```

```python
#Load in Disconnected File
disconnected_file = "../sortlog/h3c-wx7-20161118/h3c-wx7-20161118-info-formatted/h3c-wx7-20161118-disconnected-formatted.csv"
discon_df = pd.read_csv(disconnected_file, engine='python')
discon_df.info()
```

```python
#Load in UserLoggedOff File
logged_file = "../sortlog/h3c-wx7-20161118/h3c-wx7-20161118-info-formatted/h3c-wx7-20161118-logged-formatted.csv"
logged_df = pd.read_csv(logged_file, engine='python')
```

```python
#Load in Terminated File
terminated_file = "../sortlog/h3c-wx7-20161118/h3c-wx7-20161118-info-formatted/h3c-wx7-20161118-was,terminated-formatted.csv"
term_df = pd.read_csv(terminated_file, engine='python')
```

```python
#Load in SecureLogoff File
securelog_file = "../sortlog/h3c-wx7-20161118/h3c-wx7-20161118-info-formatted/h3c-wx7-20161118-hh3cSecureLogoff-formatted.csv"
securelog_df = pd.read_csv(securelog_file, engine='python')
```

```python
#Outer join all 4 files related to disconnection
discon_df=discon_df.drop('StatusCode',axis=1)
logged_df=logged_df.drop('StatusCode',axis=1)
joined_df = pd.merge(discon_df, logged_df, how='outer', on=['UserMacAddress','Time'])
joined_df.sort_values(by="Time")
```

```python
securelog_df=securelog_df.drop('StatusCode',axis=1)
joined_df = pd.merge(joined_df, securelog_df, how='outer', on=['UserMacAddress','Time','UserName','VLANID'])
joined_df.sort_values(by="Time")
```

```python
term_df=term_df.drop('StatusCode',axis=1)
joined_df = pd.merge(joined_df, term_df, how='outer', on=['UserMacAddress','Time','UserName','VLANID'])
joined_df.sort_values(by="Time")
joined_df.head(20)
```

```python
#Output the joined dataframe to csv
joined_df.to_csv(r"../sortlog/h3c-wx7-20161118/h3c-wx7-20161118-info-formatted/h3c-wx7-20161118-disconnection-formatted.csv", index=None,header=True, na_rep='NaN')
```
